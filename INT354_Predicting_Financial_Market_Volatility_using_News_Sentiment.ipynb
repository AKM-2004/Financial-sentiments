```
## About Dataset here we are having 27 coloumns and 4101 rows
## Out of 27  -> 25 are for my top news ,1 is representing the date ,1 is representing on that day my graph was up or down
## we are having news of 16 years approx from 2000-01-03 till 2016-07-01
```

```
train = df[df["Date"]<'20150101']
test = df[df["Date"]>'20141231']
```

```
test.shape
```

```
## Now we will Do preprocessing and Training
## step 1: remove all the puchuations and will convert to lower case
## step 2: now combine all the news
## step 3: convert  news to vector Techniques to be use is (CountVectorize(This is feature extraction Techniques))
## step 4: use ML algorithmns to predict  Like(Naive Bayes, RandomForest, KNNClasiifier)
## step 5: To check the performance we will different types of metrices
```

```
## step1
## for train
data_train = train.iloc[:,2:]
data_train.replace("[^a-zA-Z]"," ",regex=True, inplace=True) #regex is true bocz we want to repalce except some data
## for test
data_test = test.iloc[:,2:]
data_test.replace("[^a-zA-Z]"," ",regex=True, inplace=True)
```

```
## step 2:
## for train
for i in data_train.columns:
  data_train[i] = data_train[i].str.lower()

## for test
for i in data_test.columns:
  data_test[i] = data_test[i].str.lower()
```

```
## step 3

## For Train
combine_train = []

for i in range(0,data_train.shape[0]):
  combine_train.append(' '.join(str(x) for x in data_train.iloc[i,:]))

## For Test
combine_test = []

for i in range(0,data_test.shape[0]):
  combine_test.append(' '.join(str(x) for x in data_test.iloc[i,:]))
```

```
combine_train[0]
```

```
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

## For Train
countvec = CountVectorizer(ngram_range=(2,2))
train_data=countvec.fit_transform(combine_train)
test_data = countvec.transform(combine_test)
```

```
# For RandomForest
model_1 = RandomForestClassifier(n_estimators=150,criterion="entropy")
model_1.fit(train_data,train["Label"])
```

```
x_pred = model_1.predict(test_data)
print("This is for RandomForest")
print(confusion_matrix(x_pred,test["Label"]))
print(accuracy_score(x_pred,test["Label"])*100)
print(classification_report(x_pred,test["Label"]))
```

```
## For Naivebayes

model_2 = MultinomialNB()
model_2.fit(train_data,train["Label"])

x_pred = model_2.predict(test_data)
print("This is for NaiveBayes")
print(confusion_matrix(x_pred,test["Label"]))
print(accuracy_score(x_pred,test["Label"])*100)
print(classification_report(x_pred,test["Label"]))
```

```
## For KNN classifier

model_3 = KNeighborsClassifier(n_neighbors=5)
model_3.fit(train_data,train["Label"])

x_pred = model_3.predict(test_data)
print("This is for KNNClassifier")
print(confusion_matrix(x_pred,test["Label"]))
print(accuracy_score(x_pred,test["Label"])*100)
print(classification_report(x_pred,test["Label"]))
```

```
## For DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier
model_4 = DecisionTreeClassifier()
model_4.fit(train_data,train["Label"])

x_pred = model_4.predict(test_data)
print("This is for KNNClassifier")
print(confusion_matrix(x_pred,test["Label"]))
print(accuracy_score(x_pred,test["Label"])*100)
print(classification_report(x_pred,test["Label"]))
```
